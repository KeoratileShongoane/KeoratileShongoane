```r
 data <- read_csv("/content/API_EN.ATM.CO2E.PC_DS2_en_csv_v2_3403796.csv")
```

```r
head(data)
```

```r
# et column names from the second row and remove the first two rows
colnames(data) <- data[2, ]
data <- data[-c(1, 2), ]
head(data)
```

```r
data_separated <- data %>% separate( "Indicator Name,Indicator Code,1960,1961,1962,1963,1964,1965,1966,1967,1968,1969,1970,1971,1972,1973,1974,1975,1976,1977,1978,1979,1980,1981,1982,1983,1984,1985,1986,1987,1988,1989,1990,1991,1992,1993,1994,1995,1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022,2023,",
        into = c("Indicator Name", "Indicator Code", as.character(1960:2023)),
         sep = ",",

        fill = "right")
```

```r
head(data_separated)
```

```r
# Identify the year columns from 1960 to 2023
year_cols <- as.character(1960:2023)


# Reshape the data from wide to long format
data_long <- data_separated %>%
  pivot_longer(
    cols = all_of(year_cols),
    names_to = "Year",
    values_to = "CO2_Emissions"
  ) %>%
  mutate(
    Year = as.numeric(Year),
    CO2_Emissions = as.numeric(CO2_Emissions)
  )
```

```r
head(data_long)
```

```r
# Filtered data
filtered_data <- data_long %>%
  filter(Year >= 2010 & Year <= 2020)
```

```r

# Further filter for specific countries
countries_to_filter <- c("Ethiopia", "Mali", "Kenya")
filtered_data <- filtered_data %>%
     filter(`Country Name` %in% countries_to_filter)
```

```r
# Define custom color palette (example)
blue_palette <- c("#1f77b4", "#ff7f0e", "#2ca02c")  # Adjust as needed

ggplot(filtered_data, aes(x = Year, y = CO2_Emissions, color = `Country Name`, group = `Country Name`)) +
  geom_line(size = 1) +                                  # Draw lines for CO2 emissions trends, with specified line thickness
  geom_point(aes(shape = `Country Name`), size = 3) +   # Add points with different shapes per country
  scale_color_manual(values = blue_palette) +            # Apply custom color palette to countries
  scale_shape_manual(values = c(16, 17, 18)) +           # Define shapes: 16 (circle), 17 (triangle), 18 (square)
  labs(
    title = "CO2 Emissions Trends (2010-2020)",         # Title for the plot
    x = "Year",                                          # Label for x-axis
    y = "CO2 Emissions (metric tons per capita)"        # Label for y-axis
  ) +
  scale_x_continuous(
    breaks = seq(2010, 2020, by = 1),                    # Set breaks on the x-axis (every year)
    limits = c(2010, 2020)                               # Set limits for the x-axis (2010 to 2020)
  ) +
  theme_classic() +                                     # Use a clean, minimal theme for the plot
  theme(
    panel.grid.major = element_line(color = "grey", size = 0.2),  # Add major grid lines
    panel.grid.minor = element_line(color = "grey", size = 0.1),  # Add minor grid lines
    legend.position = "right"                            # Position legend to the right
  )
```

```r
# Correct country names in the dataset
filtered_data_corrected <- data_long %>%
  mutate(
    `Country Name` = case_when(
      `Country Name` == "Democratic Republic of the Congo" ~ "Democratic Republic of the Congo",
      `Country Name` == "Republic of Congo" ~ "Republic of the Congo",
      `Country Name` == "Egypt" ~ "Egypt",
      `Country Name` == "Ivory Coast" ~ "Côte d'Ivoire",
      `Country Name` == "Western Sahara" ~ "Western Sahara",
      TRUE ~ `Country Name`
    )
  )

# Filter the corrected data for the year 2020
data_2020_corrected <- filtered_data_corrected %>%
  filter(`Country Name` %in% african_countries & Year == 2020)

# Join the corrected data with map data
map_data_corrected <- joinCountryData2Map(data_2020_corrected, joinCode = "NAME", nameJoinColumn = "Country Name")
```

```r
# Get world map data
world <- ne_countries(scale = "medium", returnclass = "sf")

# Filter for African countries
african_map <- world %>%
  filter(continent == "Africa")

# Correct country names and merge with map data
data_2020_sf <- african_map %>%
  left_join(data_2020_corrected, by = c("name" = "Country Name"))

# Plot the map
ggplot(data_2020_sf) +
  geom_sf(aes(fill = CO2_Emissions)) +
  scale_fill_viridis_c(name = "CO2 Emissions") +
  labs(title = "CO2 Emissions in Africa (2020)") +
  theme_minimal()
```

```r
# Load necessary libraries
library(ggplot2)
library(dplyr)

# Define Southern African countries
southern_african_countries <- c("South Africa", "Namibia", "Botswana", "Zimbabwe",
                                 "Lesotho", "Swaziland", "Angola", "Mozambique",
                                 "Zambia", "Malawi")

# Filter the data for Southern African countries and the year 2020
data_southern_africa <- filtered_data %>%
  filter(`Country Name` %in% southern_african_countries & Year == 2020)

# Sort countries based on CO2 emissions
data_southern_africa_sorted <- data_southern_africa %>%
  arrange(desc(CO2_Emissions))  # Sort in descending order

# Create a bar chart to visualize the ranking of countries by CO2 emissions
ggplot(data_southern_africa_sorted, aes(x = reorder(`Country Name`, CO2_Emissions), y = CO2_Emissions, fill = `Country Name`)) +
  geom_bar(stat = "identity") +
  coord_flip() +  # Flip the coordinates for a horizontal bar chart
  labs(
    title = "Ranking of African Countries by CO2 Emissions (2020)",
    x = "Country",
    y = "CO2 Emissions (metric tons per capita)"
  ) +
  theme_minimal() +
  scale_fill_viridis_d()  # Use a color palette suitable for rankings
```
